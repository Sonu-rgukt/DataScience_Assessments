```python
data=pd.read_csv("employeedata.csv")
data
```

```python
data['Experience_Years'] = data['Experience_Years'].fillna(data['Experience_Years'].mean())
data['Salary'] = data['Salary'].fillna(data['Salary'].mean())
data
```

```python
x=np.array(data["Experience_Years"])
y=np.array(data["Salary"])
b0=float(input("enter b0 value:"))
b1=float(input("enter b1 value:"))
e=int(input())
y_pred=(b0+(b1*x)+e)
plt.scatter(x,y,c="b")
plt.plot(x,y_pred,c="r")

plt.show()
```
